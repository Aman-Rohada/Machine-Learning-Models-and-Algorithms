```python
#Removing the Null values from the data.
df = df[pd.notnull(df['CustomerID'])]
```

```python
#Excluding Negative Values in Quantity and UnitPrice
df = df[(df.Quantity>0) & (df.UnitPrice>0)]
```

```python
#Returned items indicated with C so we can drop them by filtering 
df = df[~df['StockCode'].str.contains('C')]
```

```python
#Dropping Duplicated Values
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.Country.value_counts(normalize=True).plot(kind='bar')
```

```python
df = df[df.Country == 'United Kingdom']
```

```python
print('Orders from {} to {}'.format(df['InvoiceDate'].min(),
                                    df['InvoiceDate'].max()))
```

```python
df["Total Revenue"] = df['Quantity']*df['UnitPrice']
Monetary = df.groupby('CustomerID')['Total Revenue'].sum()
Monetary = Monetary.reset_index()
Monetary.head()
```

```python
Frequency = df.groupby('CustomerID')['InvoiceNo'].count()
Frequency = Frequency.reset_index()
Frequency.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceYearMonth'] = df['InvoiceDate'].map(lambda date: 100*date.year + date.month)
df['Date'] = df['InvoiceDate'].dt.strftime('%Y-%m')
df['diff'] = max(df['InvoiceDate'])-df['InvoiceDate']
Recency = df.groupby('CustomerID')['diff'].min()

Recency = Recency.reset_index()
Recency.head()
```

```python
Recency['diff'] = Recency['diff'].dt.days
Recency.head()
```

```python
rfm = pd.merge(Recency, Frequency, on = "CustomerID", how = "inner")
```

```python
rfm = pd.merge(rfm, Monetary, on="CustomerID", how = "inner")
```

```python
rfm.columns = ["CustomerID", 'Recency', 'Frequency', 'Monetary']
```

```python
rfm
```

```python
rfm["R"] = pd.qcut(rfm["Recency"],5,labels=[5,4,3,2,1])
rfm["F"] = pd.qcut(rfm["Frequency"],5,labels=[1,2,3,4,5])
rfm["M"] = pd.qcut(rfm["Monetary"],5,labels=[1,2,3,4,5])
rfm["RFM_Score"] = rfm["R"].astype(str) +rfm["F"].astype(str) + rfm["M"].astype(str)
```

```python
rfm
```

```python
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[3-4][4-5]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm['Segment'] = rfm['R'].astype(str) + rfm['F'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm.head()
```

```python
rfm.groupby('Segment').mean().sort_values('Monetary')
```

```python
file_name = 'rfm_df1.xlsx'
rfm.to_excel(file_name)
print('DataFrame is written to Excel')
```
