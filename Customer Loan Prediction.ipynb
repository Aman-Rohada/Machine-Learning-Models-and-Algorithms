# 2.TWO SAMPLE INDEPENDENT T-TEST

```python
Married_Person = data1[(data1['Married'] == 'Yes')]
Married_Person.reset_index(inplace= True)
Married_Person
```

```python
NotMarried_Person = data1[(data1['Married'] == 'No')]
NotMarried_Person.reset_index(inplace= True)
NotMarried_Person
```

# Hypothesis
* H_0: There is no significant difference between Applicant Income whether they are married or not.

* H_1: There is significant difference between Applicant Income whether they are married or not.

* Step 2 = Two Tailed

* Step 3 = Alpha = 0.05



```python
import scipy
twosample_results =ttest_ind(Married_Person['ApplicantIncome'],NotMarried_Person['ApplicantIncome'])    #ind=independent
twosample_results
```

# Conclusion
* Since p-value is greter than level of significace(0.05),We do not Reject Ho.
* Therefore there is no significant difference between Applicant Income whether they are married or not.
