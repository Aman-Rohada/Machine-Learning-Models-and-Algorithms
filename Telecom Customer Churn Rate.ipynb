# 2.TWO SAMPLE T-TEST

# Hypothesis

* H0 : There is no difference in job satisfaction between employees with a high work-life balance(WorkLifeBalance=4) and                employees with a low work-life balance(WorkLifeBalance=1)
* H1 : There is difference in job satisfaction between employees with a high work-life balance(WorkLifeBalance=4) and employees        with a low work-life balance(WorkLifeBalance=1)

```python
wor_bal_1 = data[data['WorkLifeBalance']== 1]
wor_bal_1
```

```python
wor_bal_4 = data[data['WorkLifeBalance']== 4]
wor_bal_4
```

```python
mean_jobsat_wlbalance1 = data[data['WorkLifeBalance'] == 1]['JobSatisfaction'].mean()
mean_jobsat_wlbalance1
```

```python
mean_jobsat_wlbalance4 = data[data['WorkLifeBalance'] == 4]['JobSatisfaction'].mean()
mean_jobsat_wlbalance4
```

```python
jobsat_wlbalance1 = data[data['WorkLifeBalance'] == 1]['JobSatisfaction']
jobsat_wlbalance4 = data[data['WorkLifeBalance'] == 4]['JobSatisfaction']
t_stat, p_value = ttest_ind(jobsat_wlbalance1, jobsat_wlbalance4)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

# Conclusion

* Since p-value = 0.535, which is gretaer than the level of significance(0.05) we fail to reject H0.
* Hence we conclude there is no difference in job satisfaction between employees with a high work-life           balance(WorkLifeBalance=4) and employees with a low work-life balance(WorkLifeBalance=1) 

# 3.PEARSON'S CORRELATION

# ASSUMPTIONS

* Observations in each sample are independent and identically distributed (iid).
* Observations in each sample are normally distributed. 
* Observations in each sample have the same variance.

# Hypothesis

* H0 : There is no significant correlation between Age and Job Satisfaction.
* H1 : There is significant correlation between Age and Job Satisfaction.


```python
from scipy.stats import pearsonr
age = data['Age']
JobSat = data['JobSatisfaction']
# TO calculate the correlation coefficient and p-value
corr, p_value = pearsonr(age, JobSat)
print("Correlation coefficient:", corr)
print("P-value:", p_value)
```

# Conclusion

* As Correlation coefficient: -0.0049 which is approximately equal to 0, so we can say that there is no correlation between age   and Job satisfaction
* Since p-value = 0.851 is gretaer than level of significance(0.05) we fail to reject H0.
* Therefore we can conclude that, there is no significant correlation between Age and Job Satisfaction.
